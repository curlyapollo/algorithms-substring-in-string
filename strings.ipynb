```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv("/content/10000symb_bin_zeroWildcard.csv", sep=';')

for el in df1.columns:
    if (el != "Sample size"):
        plt.plot(df1["Sample size"], df1[el], label=el)
    
plt.legend()
plt.xlabel('Sample size')
plt.ylabel('Time, ns')
plt.title('Текст размера 10000, алфавит из двух элементов, 0 символов подстановки.\n Размер шаблона 100-3000')


plt.show()
```
